```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a figure and subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Plot the first histogram
sns.histplot(data = df_partA_train,x = 'Density_Count', bins=20, color='blue', alpha=0.7,kde=True,hue='Channel_Name', ax=axs[0, 0])
axs[0, 0].set_xlabel('Crowd Counts')
axs[0, 0].set_ylabel('Frequency')
axs[0, 0].set_title('Histogram of Crowd Counts df_partA_train')
axs[0, 0].grid(True)

# Plot the second histogram
sns.histplot(data = df_partA_test,x = 'Density_Count', bins=20, color='red', alpha=0.7, kde=True,hue='Channel_Name', ax=axs[0, 1])
axs[0, 1].set_xlabel('Crowd Counts')
axs[0, 1].set_ylabel('Frequency')
axs[0, 1].set_title('Histogram of Crowd Counts df_partA_test')
axs[0, 1].grid(True)

# Plot the third histogram
sns.histplot(data = df_partB_train, x='Density_Count', bins=20, color='green', alpha=0.7, kde=True,hue='Channel_Name', ax=axs[1, 0])
axs[1, 0].set_xlabel('Crowd Counts')
axs[1, 0].set_ylabel('Frequency')
axs[1, 0].set_title('Histogram of Crowd Counts df_partB_train')
axs[1, 0].grid(True)

# Plot the fourth histogram
sns.histplot(data = df_partB_test,x = 'Density_Count', bins=20, color='purple', alpha=0.7, kde=True,hue='Channel_Name',ax=axs[1, 1])
axs[1, 1].set_xlabel('Crowd Counts')
axs[1, 1].set_ylabel('Frequency')
axs[1, 1].set_title('Histogram of Crowd Counts df_partB_test')
axs[1, 1].grid(True)

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
