```python
part_A_train = os.path.join(os.getcwd(),'ShanghaiTech\\part_A\\train_data\\images')
part_A_test = os.path.join(os.getcwd(),'ShanghaiTech\\part_A\\test_data\\images')
part_B_train = os.path.join(os.getcwd(),'ShanghaiTech\\part_B\\train_data\\images')
part_B_test = os.path.join(os.getcwd(),'ShanghaiTech\\part_B\\test_data\\images')
```

```python
part_A_train_images = glob.glob(os.path.join(part_A_train,'*.jpg'))
part_A_test_images = glob.glob(os.path.join(part_A_test,'*.jpg'))
part_B_train_images = glob.glob(os.path.join(part_B_train,'*.jpg'))
part_B_test_images = glob.glob(os.path.join(part_B_test,'*.jpg'))
print(f'Shanghai Part A train -- total size {len(part_A_train_images)}')
print(f'Shanghai Part A test -- total size {len(part_A_test_images)}')
print(f'Shanghai Part B train -- total size {len(part_B_train_images)}')
print(f'Shanghai Part B test -- total size {len(part_B_test_images)}')
print(f'Total people counting images = {len(part_A_train_images)+len(part_A_test_images)+len(part_B_train_images)+len(part_B_test_images)}')
```

```python
def merging_all_images(list_of_images,list_names):
    
    final_paths = []
    
    for index,(path,name) in enumerate(zip(list_of_images,list_names),0):
        
        def extract_number(path):
            
            file_path = path.split('\\')[-1].split('.')[0].split('_')[1]
            
            return int(file_path)

        file_paths = sorted(path,key=extract_number)
        
        final_paths.append({name:file_paths})
        
    return final_paths
```

```python
list_names = ["part_A_train_images","part_A_test_images","part_B_train_images","part_B_test_images"]
list_of_images_path = [part_A_train_images,part_A_test_images,part_B_train_images,part_B_test_images]
```

```python
all_images_files = merging_all_images(list_of_images_path,list_names)
len(all_images_files)
```

```python
for i in list_of_images_path:
    print(len(i))
```

```python
import os

def changing_images_sequence(list_of_paths):
    start_number = 1
    new_list = [{"part_A_train_images":[]},{"part_A_test_images": []}, {"part_B_train_images": []}, {"part_B_test_images": []}]
    
    for i in range(len(list_of_paths)):
        stop_name = list(list_of_paths[i].keys())[0]
        stop_number = len(list_of_paths[i][stop_name])
        
        for index, file_path in enumerate(list_of_paths[i][stop_name], start_number):
            parts = file_path.rsplit('IMG_', 1)
            prefix = parts[0]
            extension = parts[-1].split('.')[-1]
            new_file_path = f"{prefix}IMG_{index}.{extension}"
            
            # Rename the original image file
            os.rename(file_path, new_file_path)
            
            # Update the new file path in new_list
            new_list[i][stop_name].append(new_file_path)
        
        # Update start_number for the next iteration
        start_number += stop_number
    
    return new_list

# Example usage
a = changing_images_sequence(all_images_files)
```
